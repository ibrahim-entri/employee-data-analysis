## Stage 2 – Data Cleaning and Pre-processing

- Handle missing values (impute or drop)

- Handle duplicates
- column name rename or dropping unwanted columns if required

- Treat outliers if required

- Check skewness and apply transformations

- Convert data types if needed

- Feature transformations (date parts, derived fields if required for analysis)

```python
# Missing Values

df.isnull().sum()
```

```python
# Coloumn Name

print(df.columns)
```

```python
# Missing Values
# Numerical Columns

num_cols = ['Bonus %', 'Monthly Incentive',
             'Tax Deduction', 'Overtime Hours']

for col in num_cols:
  df[col] = pd.to_numeric(df[col],errors='coerce')
  df[col] = df[col].fillna(df[col].median())
```

```python
# Missing Values
# Categorical Columns

cat_cols = ['Gender','Team','Active',
            'WFH','Compliance Trained','Previous Employer']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Remove Duplicates
# there are zero duplicates

df.drop_duplicates(inplace=True)
```

```python
# Rename 'Bonus %' to 'Bonus Percentage'


df.rename(columns={'Bonus %': 'Bonus Percentage'}, inplace=True)

print(df.columns)
```

```python
# Drop Unwanted Columns

df.drop('Device MAC Address',axis=1,inplace=True)
```

```python
df['Probation End Date'].head(10)
```

```python
# Convert Data Types

df['Start Date'] = pd.to_datetime(
    df['Start Date'],
    format='mixed',
    errors='coerce'
)

df['Probation End Date'] = pd.to_datetime(
    df['Probation End Date'],
    format='mixed',
    errors='coerce'

)
```

```python
df[['Start Date','Probation End Date']].info()
```

```python
print(df.columns)
```

```python
# Data Type Check

df[['Start Date','Probation End Date']].info()
```

```python
# Checking missing Values

df[['Start Date','Probation End Date']].isnull().sum()
```

 Observation:

Start Date was successfully converted with no missing values.

Probation End Date contains 46 missing/invalid values (NaT).

These values were retained because no reliable business rule existed for imputation.

```python
# Feature transformations (date parts, derived fields if required for analysis)

# Joing Year

df['Joining Year'] = df['Start Date'].dt.year

df[['Start Date','Joining Year']].head()

```

```python
# Joing Month

df['Joining Month'] = df['Start Date'].dt.month
df[['Start Date','Joining Year','Joining Month']].head()
```

```python
# Probation Duration

df['Probation Days'] = (
    df['Probation End Date'] -
    df['Start Date']
).dt.days
df[['Probation Days']].head()
```

```python
# Stage 3 – EDA and Visualizations
# Univariate Analysis (Single Variable)

# Gender Distribution
sns.countplot(x='Gender',data=df)
plt.title('Gender Distribution')
plt.show()
```

The countplot displays the distribution of employees by gender.

```python
# Team Distribution

sns.countplot(
    y='Team',
    data=df,
    order=df['Team'].value_counts().index
)
plt.title('Employee Distribution by Team')
plt.show()
```

**Team Distribution Interpretation**

The chart shows the distribution of employees across departments. Client Services has the highest number of employees, while Legal has the lowest. Most departments have a relatively balanced number of employees.

```python
# Salary Distribution

plt.hist(df['Salary'],bins=25)

plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Employee Count')

plt.show()
```

**Salary Distribution Interpretation**

The histogram shows the distribution of employee salaries.  
Most employees fall within the mid-range salary brackets, while fewer employees earn very high or very low salaries.  
This indicates that the organization’s salaries are concentrated around typical ranges, with only a small proportion of employees earning at the extremes.


```python
# Salary Boxplot

plt.boxplot(df['Salary'])
plt.title('Salary Distribution')
plt.show()
```

**Salary Boxplot**

The boxplot shows the distribution of employee salaries and highlights outliers.  
The median salary is around ₹90,000, with most employees earning between ₹60,000 and ₹120,000.  
A few employees fall outside this range, indicating outliers with unusually high or low salaries.  
This analysis helps identify salary variability and potential disparities within the organization.


```python
# WFH Distribution

sns.countplot(
    x=('WFH'),
    data=df,
    order=df['WFH'].value_counts().index
)

plt.title('WFH Distribution')
plt.xticks(rotation=45)
plt.show()
```

###** Interpretation**
The countplot shows the distribution of employees by work-from-home status.  
Most employees are in the **Pending** category, while the remaining categories (Hybrid, Yes, No, Remote Only) are relatively balanced.  
This suggests that the organization has a diverse mix of work arrangements, with many employees still awaiting confirmation of their WFH status.


```python
# Bivariate Analysis (Two Variables)

sns.boxplot(x='Gender',y='Salary',data=df)
plt.title('Salary Distribution by Gender')
plt.show()
```

### Interpretation
The boxplot compares salary distribution across genders.  
It shows that **male employees have slightly higher salaries than female employees**.  
The difference is small, but it indicates a minor variation in pay between genders.



```python
# Team vs Salary (Two Variables)

sns.boxplot(
    x='Team',
    y='Salary',
    data=df,
    order=df.groupby('Team')['Salary'].mean().sort_values(ascending=False).index
)
plt.title('Salary Distribution by Team')
plt.xticks(rotation=45)
plt.show()
```

**Salary Distribution by Team Interpretation**

The boxplot compares salary distributions across teams, ordered by average salary.  
Engineering, Finance, and Sales show higher median salaries, while teams like Distribution and Client Services have lower ranges.  
The chart also highlights variability within teams, with some departments showing wide salary spreads and outliers.  
This analysis helps identify both overall pay differences between teams and internal disparities within each department.


```python
# Joining Year Trend (Two Variables)

df['Joining Year'].value_counts().sort_index().plot()

plt.title('Employee Joining Trend')
plt.xlabel('Year')
plt.ylabel('Employee Count')
plt.show()
```

### Interpretation
The line chart shows the trend of employee Joining Trend.  
Recruitment levels fluctuate over time, with noticeable peaks around 1995 and 2010, and dips around 1990 and 2018.  
This highlights periods of workforce expansion and slowdown, giving insights into the organization’s hiring patterns across decades.


```python
# WFH vs Salary

sns.barplot(
    x='WFH',
    y='Salary',
    data=df
)

plt.xticks(rotation=45)
plt.title('Salary Distribution by WFH Status')
plt.show()
```

```python
# Multivariate Analysis (3+ Variables)

pivot = pd.pivot_table(
    df,
    index='Team',
    values='Salary',
    columns='Gender',
    aggfunc='mean'
)
pivot
```

```python
sns.barplot(
    x='Team',
    y='Salary',
    hue='Gender',
    data=df
)
plt.xticks(rotation=45)
plt.title('Average Salary by Team and Gender')
plt.show()
```

### Interpretation
The pivot table compares average salaries across teams by gender.  
It shows that salary differences vary by department:
- In Business Development, Legal, and Marketing, **female employees earn slightly higher average salaries**.  
- In Client Services, Distribution, Engineering, Finance, and Sales, **male employees earn higher average salaries**.  
- In Human Resources and Product teams, salaries are **nearly equal**.  

Overall, the organization has a **mixed pattern of salary distribution**, with no single gender consistently earning more across all teams.


```python
# Correlation Heatmap

numeric_cols = [
    'Salary',
    'Bonus Percentage',
    'Monthly Incentive',
    'Tax Deduction',
    'Overtime Hours'
]

corr = df[numeric_cols].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    vmin=-1,
    vmax=1
)

plt.title('Correlation Heatmap')
plt.show()
```

**Interpretation**

The correlation heatmap shows that the financial and work-related factors in the dataset are largely independent. Salary does not strongly correlate with bonus percentage, incentives, tax deductions, or overtime hours. Similarly, other variables show only minimal relationships.

This suggests that each factor operates separately, and changes in one (like overtime hours) do not significantly affect another (like salary or bonus). The organization’s compensation structure appears to treat these elements as distinct components rather than tightly linked measures.
